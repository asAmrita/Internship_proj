```python
import pandas as pd

# Load the dataset
mcdonalds = pd.read_csv("/content/mcdonalds.csv")

# Get the column names of the dataset
column_names = mcdonalds.columns

print(column_names)
```

```python
import numpy as np
MD_x = mcdonalds.iloc[:, 0:11]

# Convert "Yes" to 1 and other values to 0
MD_x = (MD_x == "Yes").astype(int)

# Calculate the mean of each column and round to 2 decimal places
#col_means = MD_x.mean().round(2)
rounded_means = np.round(np.mean(MD_x, axis=0), 2)

print(rounded_means)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
data_scaled = scaler.fit_transform(MD_x)

pca = PCA()
data_pca = pca.fit_transform(data_scaled )
print(data_pca)
```

```python
# Summary of PCA
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

components = pca.components_

# Display results
print("Explained Variance by each component:\n", explained_variance)
print("\nCumulative Variance explained:\n", cumulative_variance)
print("\nPrincipal Components:\n", components)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming MD.x is a DataFrame or a NumPy array
# MD_x = ...

# Set the seed for reproducibility
np.random.seed(1234)

def bootFlexclust(data, k_range, nrep=10, nboot=100):
    results = {k: [] for k in k_range}
    n_samples = data.shape[0]

    for k in k_range:
        ari_scores = []
        for _ in range(nboot):
            # Bootstrap resampling
            sample_indices = np.random.choice(n_samples, n_samples, replace=True)
            boot_sample = data[sample_indices]
            original_sample = data

            # Fit KMeans to the bootstrap sample
            kmeans_boot = KMeans(n_clusters=k, random_state=1234).fit(boot_sample)
            kmeans_orig = KMeans(n_clusters=k, random_state=1234).fit(original_sample)

            # Calculate ARI
            ari = adjusted_rand_score(kmeans_orig.labels_, kmeans_boot.labels_)
            ari_scores.append(ari)

        results[k] = ari_scores

    return results

# Running the bootstrap clustering
MD_b28 = bootFlexclust(MD_x, range(2, 9), nrep=10, nboot=100)

# Preparing data for plotting
mean_ari = {k: np.mean(v) for k, v in MD_b28.items()}
ari_df = pd.DataFrame.from_dict(mean_ari, orient='index', columns=['ARI'])

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=ari_df, marker='o')
plt.xlabel('Number of Segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Adjusted Rand Index vs. Number of Segments')
plt.show()
```
